```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("derived_datasets/final_customer_data_11.csv")
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f"Duplicates found: {duplicates}")
```

```python
# Check categorical values
print("\nUnique values in categories:")
print("Social Media:", data['social_media_platform'].unique())
print("Sentiment:", data['review_sentiment'].unique())
```

```python
# Check if social media matches customer ID
missing_id = data[(data['social_media_platform'] != 'Unknown') & (data['customer_id_new'].isna())].shape[0]
print(f"Transactions with missing customer_id_new but known platform: {missing_id}")
```

#### 2. Statistical Summarization

```python
# Summary of numbers
print("\nSummary of numerical columns:")
print(data[['purchase_amount', 'customer_rating', 'engagement_score']].describe())

# Plot purchase amount before and after simple augmentation
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
data['purchase_amount'].hist(bins=20)
plt.title('Purchase Amount Before')
plt.xlabel('Amount')
plt.ylabel('Count')

# Add simple noise for demo
data['purchase_amount_aug'] = data['purchase_amount'] * (1 + 0.1 * (np.random.randn(len(data))))
plt.subplot(1, 2, 2)
data['purchase_amount_aug'].hist(bins=20)
plt.title('Purchase Amount After')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 3. Feature Selection
# Simple correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data[['purchase_amount', 'customer_rating', 'engagement_score',
                'purchase_interest_score', 'customer_engagement_score']].corr(),
            annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pick top features (manual selection for simplicity)
top_features = ['purchase_amount', 'customer_rating', 'engagement_score',
                'purchase_interest_score', 'customer_engagement_score',
                'monthly_total_spend', 'monthly_avg_spend', 'avg_spent_last_3',
                'transaction_id', 'customer_id_legacy']
final_df = data[top_features + ['purchase_date', 'social_media_platform', 'review_sentiment']]

# 4. Save Final Dataset
group_number = "11"
final_df.to_csv(f'derived_datasets/final_dataset_ready_{group_number}.csv', index=False)
print(f"\nSaved final dataset as 'final_dataset_ready_{group_number}.csv'")
print(f"Final dataset size: {final_df.shape}")
```
